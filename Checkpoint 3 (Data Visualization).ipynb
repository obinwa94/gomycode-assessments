```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sb.diverging_palette( 220 , 10 , as_cmap = True )

    s = sb.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 15 }

        )
plot_correlation_map(titanic)
```

```python
t2 = titanic.groupby(['Survived', 'Pclass']).agg(total =('Survived', 'count'))
t2
```

```python
list1 = ["{:,.2%}".format(r/t2['total'][:3].sum()) for r in t2['total'][:3]]
list1
```

```python
list2 = ["{:,.2%}".format(r/t2['total'][3:].sum()) for r in t2['total'][3:]]
list2
```

```python
list3 = list1 + list2 #Merging list1 & list2 to give a new list
list3
```

```python
list1.extend(list2) #Merging list2 to list1
list1
```

```python
t2['Proportion'] = list1 #Adding column to t2
t2
```

```python
td = {

                "Capt":       "Officer",

                "Col":        "Officer",

                "Major":      "Officer",

                  "Dr":         "Officer",

                "Rev":        "Officer",

                "Jonkheer":   "Royalty",

                "Don":        "Royalty",

                "Sir" :       "Royalty",

               "Lady" :      "Royalty",

              "the Countess": "Royalty",

                "Dona":       "Royalty",

                "Mme":        "Miss",

                "Mlle":       "Miss",

                "Miss" :      "Miss",

                "Ms":         "Mrs",

                "Mr" :        "Mrs",

                "Mrs" :       "Mrs",

                "Master" :    "Master"

                }
td
```

```python
td.keys() #Calling out dictionary keys
```

```python
td.values() #calling out dictionary values
```

```python
old_title = []
new_title = []
for name in titanic['Name']:
    for t,u in zip(td.keys(), td.values()):
        t = t + "."
        if t in name:
            old_title.append(t)
            new_title.append(u)
```

```python
len(old_title) == len(titanic) #comparing length of variables
```

```python
len(new_title) == len(titanic)#comparing length of variables
```

```python
titanic['old_title'] = old_title   #Adding new column
titanic['new_title'] = new_title
titanic
```

```python
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch']  #combining SibSp and Parch
titanic
```

```python
plt.xlabel("New title")
plt.ylabel("New title value count")
plt.title("New title relation")
bn = titanic["new_title"].value_counts()
bn.plot.bar(rot = 45)
```

```python
sb.countplot(data = titanic, x='new_title', palette = ['red','blue'], hue = 'Survived')
```
